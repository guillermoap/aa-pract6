```python
test_ann = NeuralNetwork(nin=nin, nhidden=10, nout=nout, activation=sigmoid)
test_ann.setup_gradient_check(epsilon=10**-4)
training_points = generate_points(number=500, func=func, point_size=nin)
```

```python
error = pd.DataFrame(test_ann.gradient_error, columns=['iter', 'ErrorGradiente'])
```
